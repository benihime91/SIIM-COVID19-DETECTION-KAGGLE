```python
def txt2pandas(Idxs, fold):
    image_ids = []
    PredictionStrings = []
    
    for Idx in tqdm(Idxs, dynamic_ncols=True):
        
        txt_label = os.path.join(SUB_DIR, f'fold{fold}/labels/{Idx}.txt')
        
        H,W = image_df.loc[image_df.image_id == Idx, ['dim0', 'dim1']].values[0]
        
        bboxes = []
    
        if os.path.isfile(txt_label):
            in1 = open(txt_label, 'r')
            lines = in1.readlines()
            in1.close()

            for line in lines:
                arr = line.strip().split(' ')
                class_id = 'opacity'
                x = float(arr[1])
                y = float(arr[2])
                w = float(arr[3])
                h = float(arr[4])
                x1 = x - (w / 2)  
                x2 = x + (w / 2)    
                y1 = y - (h / 2)
                y2 = y + (h / 2)
                conf = arr[5]

                x1 = int(round(x1 * W))
                y1 = int(round(y1 * H))
                x2 = int(round(x2 * W))
                y2 = int(round(y2 * H))

                bboxes.append(f"{class_id} {conf} {x1} {y1} {x2} {y2}")
        else:
            print(txt_label)
            class_id = 'none'
            conf = 1.0
            [x1, y1, x2, y2] = [0, 0, 1, 1]
            bboxes.append(f"{class_id} {conf} {x1} {y1} {x2} {y2}")

        image_ids.append(Idx)
        PredictionStrings.append(' '.join(bboxes))
    
    detections = pd.DataFrame()    
    detections['image_id'] = image_ids
    detections['image_id'] = detections['image_id'].map(lambda x: x + '_image')
    detections['PredictionString'] = PredictionStrings
    detections.to_csv(f'/kaggle/working/subm/yolo_fold{fold}.csv', index=False)
    print(f'Dataframe saved to /kaggle/working/subm/yolo_fold{fold}.csv')
```

```python
def run_wbf(subm_list, iou_same=0.6, out_path=None, skip_box_thr=0.0005):
    max_value = 10000
    preds   = []
    checker = None
    
    for path in subm_list:
        s = pd.read_csv(path)
        s.sort_values('image_id', inplace=True)
        s.reset_index(drop=True, inplace=True)
        ids = s['image_id']
        if checker:
            if tuple(ids) != checker:
                print(set(checker) - set(ids))
                print('Different IDS!', len(tuple(ids)), path)
                exit()
        else:
            checker = tuple(ids)
        preds.append(s['PredictionString'].values)

    if out_path is None:
        out_path = '/kaggle/working/' + 'ensemble_iou_{}.csv'.format(iou_same)
    
    out = open(out_path, 'w')
    
    out.write('image_id,PredictionString\n')
    
    for j, id in enumerate(tqdm(list(checker))):
        
        boxes_list = []
        scores_list = []
        labels_list = []
        empty = True
        for i in range(len(preds)):
            boxes = []
            scores = []
            labels = []
            p1 = preds[i][j]
            if str(p1) != 'nan':
                arr = p1.strip().split(' ')
                for k in range(0, len(arr), 6):
                    cls = 1 if arr[k] == 'opacity' else 0
                    prob = float(arr[k + 1])
                    x1 = float(arr[k + 2]) / max_value
                    y1 = float(arr[k + 3]) / max_value
                    x2 = float(arr[k + 4]) / max_value
                    y2 = float(arr[k + 5]) / max_value
                    boxes.append([x1, y1, x2, y2])
                    scores.append(prob)
                    labels.append(cls)

            boxes_list.append(boxes)
            scores_list.append(scores)
            labels_list.append(labels)

        boxes, scores, labels = weighted_boxes_fusion(
            boxes_list,
            scores_list,
            labels_list,
            iou_thr=iou_same,
            skip_box_thr=skip_box_thr,
            conf_type='max',
        )

        if len(boxes) == 0:
            out.write('{},none 1 0 0 1 1\n'.format(id, ))
        else:
            final_str = ''
            for i in range(len(boxes)):
                cls = int(labels[i])
                prob = scores[i]
                x1 = int(boxes[i][0] * max_value)
                y1 = int(boxes[i][1] * max_value)
                x2 = int(boxes[i][2] * max_value)
                y2 = int(boxes[i][3] * max_value)
                if cls == 0:
                    final_str += '{} {} {} {} {} {} '.format('none', prob, 0, 0, 1, 1)
                else:
                    final_str += '{} {} {} {} {} {} '.format('opacity', prob, x1, y1, x2, y2)
            out.write('{},{}\n'.format(id, final_str.strip()))

    out.close()
    print(f"Results ---> {out_path}")
    return out_path
```

```python
subm_list = [
    '/kaggle/working/subm/yolo_fold0.csv',
    '/kaggle/working/subm/yolo_fold1.csv',
    '/kaggle/working/subm/yolo_fold2.csv',
    '/kaggle/working/subm/yolo_fold3.csv',
    '/kaggle/working/subm/yolo_fold4.csv'
]

run_wbf(subm_list, out_path='/kaggle/working/yolo_v5.csv')
predictions = pd.read_csv('/kaggle/working/yolo_v5.csv', )
predictions.head()
```
